```python
#importing libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
np.random.seed(42)
%matplotlib inline
from sklearn.linear_model import LinearRegression
```

```python
# generating a data set
ndata = 100
factor = 100
x = 6*np.random.rand(ndata,1)-3
x.shape
```

```python
y = -0.5*x*x + x*2.5 + np.random.randn(ndata,1)*100
y.shape
```

```python
# creating a model of degree 1 
degree =1
poly_features = PolynomialFeatures(degree=degree, include_bias=True)
```

```python
#By adding a bias term, the model can more accurately predict outcomes because it’s no longer forced to go through the origin. This flexibility often results in better predictions, especially when the data doesn’t naturally start at zero.
```

```python
x_poly = poly_features.fit_transform(x)
```

```python
# train test split 
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_poly , y , test_size=.2,random_state=42)
```

```python
# training model via linear regression
```

```python
lin = LinearRegression()
```

```python
lin.fit(x_train, y_train)
```

```python
#calculating error metrics for training set
# first predict the values of training set 
predict_train = lin.predict(x_train)
from sklearn import metrics
```

```python
print(metrics.mean_absolute_error(y_train,predict_train))
print(metrics.mean_squared_error(y_train,predict_train))
print(metrics.r2_score(y_train,predict_train))
```

```python
#predicting for test 
predict_test = lin.predict(x_test)
print(metrics.mean_absolute_error(y_test,predict_test))
print(metrics.mean_squared_error(y_test,predict_test))
print(metrics.r2_score(y_test,predict_test))
```

```python
# for degree one as we see  vast values are generated 
#Underfitting Indicators:

#High Training Error: The model has high error on the training data.
#High Validation Error: The model also has high error on the validation data.
#Plots: Both training and validation curves will be high and might converge to similar values.

#plt.plot(x,y)
#plt.plot(x_train,predict_train)
#plt.show()
```

```python
#Overfitting Indicators:

#Low Training Error: The model has very low error on the training data.
#High Validation Error: The model has high error on the validation data.
#Plots: Training loss decreases and validation loss starts increasing after a certain point.
```
